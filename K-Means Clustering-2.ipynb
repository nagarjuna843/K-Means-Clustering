```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('C:/Users/PC/Downloads/pandas notebooks/Data sets/Salary_Data.csv')
df
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
```

```python
km=KMeans(n_clusters=4)
km
```

```python
y_predicted=km.fit_predict(df[['YearsExperience','Salary']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
df4 = df[df.cluster == 3]

plt.scatter(df1['YearsExperience'],df1['Salary'],color='green')
plt.scatter(df2['YearsExperience'],df2['Salary'],color='red')
plt.scatter(df3['YearsExperience'],df3['Salary'],color='black')
plt.scatter(df4['YearsExperience'],df4['Salary'],color='blue')

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
```

```python
df['YearsExperience'] = (df['YearsExperience']-min(df['YearsExperience']))/(max(df['YearsExperience'])-min(df['YearsExperience']))
df['Salary'] = (df['Salary']-min(df['Salary']))/(max(df['Salary'])-min(df['Salary']))
df
```

```python
km = KMeans(n_clusters=4)
y_predicted = km.fit_predict(df[['YearsExperience','Salary']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
df4 = df[df.cluster == 3]

plt.scatter(df1['YearsExperience'],df1['Salary'],color='green')
plt.scatter(df2['YearsExperience'],df2['Salary'],color='red')
plt.scatter(df3['YearsExperience'],df3['Salary'],color='black')
plt.scatter(df4['YearsExperience'],df4['Salary'],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
km.cluster_centers_
```

```python
k_range=range(1,10)
sse = []
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['YearsExperience','Salary']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_range,sse)
```
